```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer # for data prepocessing
from sklearn.naive_bayes import MultinomialNB
import seaborn as sns
```

```python
df =pd.read_csv('/kaggle/input/emails/4.emails.csv')
df
```

```python
df['spam'].value_counts()
```

```python
df.drop_duplicates(inplace = True)
df
```

```python
df['spam'].value_counts()
```

```python
df
```

```python
df.isnull().sum()
```

```python
#separate dependent and independent value
x = df.text.values
x
```

```python
y = df.spam.values
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

# data prepocessing


```python
cv = CountVectorizer()
xtrain = cv.fit_transform(xtrain)
```

```python
xtrain.toarray()
```

# ML Algo

```python
model = MultinomialNB()
model.fit(xtrain,ytrain)
```

```python
xtest = cv.transform(xtest)
```

```python
xtest.toarray()
```

```python
model.score(xtest,ytest)
```
